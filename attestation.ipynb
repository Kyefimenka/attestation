```python
import requests

url = 'https://www.cbr.ru/currency_base/daily/'
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table')
```

```python
data = []

for row in table.find_all('tr'):
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])

currency = pd.DataFrame(data)
```

```python
currency.columns = ['Цифр.код', 'Букв.код', 'Единиц', 'Валюта', 'Курс']
currency.head()
```

```python
currency = currency.drop([0]).reset_index(drop=True)
```

```python
book['Валюта'] = book['Валюта'].replace('$', 'Доллар США')
book
```

```python
book['Курс'] = 0
```

```python
book.loc[book['Валюта'] == 'Рубль', 'Курс'] = 1
book
```

```python
def substitute_exchange_rate(data_table, currency_table, currency_name):
    data_table.loc[data_table['Валюта'] == currency_name, 'Курс'] = currency_table.loc[currency_table['Валюта'] == currency_name,'Курс'].iloc[0].replace(',','.')
    return data_table
```

```python
substitute_exchange_rate(substitute_exchange_rate(book, currency, 'Доллар США'), currency, 'Евро')
```

```python
book['Маржа в рублях'] = book['Маржинальность'] * book['Цена в рублях']
```
